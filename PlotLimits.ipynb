```python
from mpl_toolkits.basemap import Basemap
from numpy import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.stats import norm
plt.rcParams['axes.linewidth'] = 2
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)


n = 100
g = logspace(-20,-6,n)
m = logspace(-12,6,n)

alph =  0.9904
C = -9.2616-0.75


QCD = zeros(shape=(n,n))
for i in range(0,n):
    m_K = 10.0**(log10(g[i])-C)/alph
    gi = 10.0**(alph*log10(m_K) + C)
    QCD[i,:] = norm.pdf(log10(m),log10(m_K),0.8);

QCD[QCD<1e-50] = 0.0
KSVZ = 10.0**(alph*log10(m) + C + 0.53)
DFSZ = 10.0**(alph*log10(m) + C + 0.1)
cols = cm.get_cmap('YlOrBr')
cols.set_under('w')
plt.pcolor(m,g,QCD,cmap=cols,vmin=1e-2,vmax=0.7,zorder=0)
#plt.pcolor(m,g,QCD,cmap=cols,vmin=1e-2,vmax=0.7,zorder=0)
#plt.pcolor(m,g,QCD,cmap=cols,vmin=1e-2,vmax=0.7,zorder=0)
plt.plot(m,KSVZ,'-',linewidth=2,color=cols(1.0),zorder=1)
plt.plot(m,DFSZ,'-',linewidth=2,color=cols(1.0),zorder=1)
plt.text(9.3466e-08,   1.2261e-15 ,r'{\bf KSVZ}',fontsize=20,rotation=50,color=cols(1.0))
plt.text( 5.2628e-07,   2.3013e-16,r'{\bf DFSZ}',fontsize=20,rotation=50,color=cols(1.0))





M87_col = [0.0, 0.66, 0.42]
M87 = loadtxt("limit_data/M87.txt")
plt.fill_between(M87[:,0],M87[:,1],y2=g[-1],edgecolor=None,facecolor=M87_col)

SNgamma_col = [0.05, 0.5, 0.06]
SNgamma = loadtxt("limit_data/SN-gamma.txt")
plt.fill_between(SNgamma[:,0],SNgamma[:,1],y2=g[-1],edgecolor=None,facecolor=SNgamma_col)

HYDRA_col = [0.24, 0.71, 0.54]
HYDRA = loadtxt("limit_data/HYDRA_A.txt")
plt.fill_between(HYDRA[:,0],HYDRA[:,1],y2=g[-1],edgecolor=None,facecolor=HYDRA_col)

HESS_col = [0.0, 0.62, 0.38]
HESS = loadtxt("limit_data/HESS.txt")
plt.fill_between(HESS[:,0],HESS[:,1],y2=g[-1],edgecolor=None,facecolor=HESS_col)

Fermi_col = [0.0, 0.42, 0.24]
Fermi1 = loadtxt("limit_data/Fermi1.txt")
Fermi2 = loadtxt("limit_data/Fermi2.txt")
plt.fill_between(Fermi1[:,0],Fermi1[:,1],y2=g[-1],edgecolor=None,facecolor=Fermi_col,alpha=0.7)
plt.fill(Fermi2[:,0],1.01*Fermi2[:,1],edgecolor=None,facecolor=Fermi_col,alpha=0.7)





XRAY_col = [0.03, 0.57, 0.82]
XRAY = loadtxt("limit_data/XRAY.txt")
plt.fill_between(XRAY[:,0],XRAY[:,1],y2=g[-1],edgecolor=None,facecolor=XRAY_col,zorder=2)

CMB_col = [0.03, 0.15, 0.4]
CMB = loadtxt("limit_data/CMB.txt")
plt.fill_between(CMB[:,0],CMB[:,1],y2=g[-1],edgecolor=None,facecolor=CMB_col,zorder=2)

EBL_col =  [0.0, 0.2, 0.6]
EBL = loadtxt("limit_data/EBL.txt")
plt.fill_between(EBL[:,0],EBL[:,1],y2=g[-1],edgecolor=None,facecolor=EBL_col,zorder=2)

x_ion_col = [0.27, 0.51, 0.71]
x_ion = loadtxt("limit_data/x_ion.txt")
plt.fill_between(x_ion[:,0],x_ion[:,1],y2=g[-1],edgecolor=None,facecolor=x_ion_col,zorder=2)

Telescopes_col = [0.09, 0.45, 0.27]
Telescopes = loadtxt("limit_data/Telescopes.txt")
plt.fill_between(Telescopes[:,0],Telescopes[:,1],y2=g[-1],edgecolor=None,facecolor=Telescopes_col,zorder=2)


HB_col = [0.0, 0.66, 0.42]
HB = loadtxt("limit_data/HorizontalBranch.txt")
plt.fill_between(HB[:,0],HB[:,1],y2=g[-1],edgecolor=None,facecolor=HB_col,zorder=2)

SolarNu_col = [0.01, 0.75, 0.24]
SolarNu = loadtxt("limit_data/SolarNu.txt")
plt.fill_between(SolarNu[:,0],SolarNu[:,1],y2=g[-1],edgecolor=None,facecolor=SolarNu_col,zorder=2)





RBF_col = [0.67, 0.31, 0.32]
RBF = loadtxt("limit_data/RBF_UF_Haloscopes.txt") 
plt.fill_between(RBF[:,0],RBF[:,1],y2=g[-1],edgecolor=None,facecolor=RBF_col,zorder=3)

ADMX_col = [0.8, 0.0, 0.0]
ADMX = loadtxt("limit_data/ADMX.txt") 
plt.fill_between(ADMX[:,0],ADMX[:,1],y2=g[-1],edgecolor=None,facecolor=ADMX_col,zorder=3)
ADMX = loadtxt("limit_data/ADMX2017.txt") 
plt.fill_between(ADMX[:,0],ADMX[:,1],y2=g[-1],edgecolor=None,facecolor=ADMX_col,zorder=3)

HAYSTAC_col= [0.88, 0.07, 0.37]
HAYSTAC = loadtxt("limit_data/HAYSTAC.txt") 
plt.fill_between(HAYSTAC[:,0],HAYSTAC[:,1],y2=g[-1],edgecolor=None,facecolor=HAYSTAC_col,zorder=3)



# HELIOSCOPES
CAST_col = [0.5, 0.0, 0.13]
CAST = loadtxt("limit_data/CAST.txt") 
plt.fill_between(CAST[:,0],CAST[:,1],y2=g[-1],edgecolor=None,facecolor=CAST_col,zorder=3)


ALPS_col = [0.8, 0.25, 0.33]
ALPS = loadtxt("limit_data/ALPS.txt") 
plt.fill_between(ALPS[:,0],ALPS[:,1],y2=g[-1],edgecolor=None,facecolor=ALPS_col,zorder=3)


# PROJECTED
IAXO_col = [0.79, 0.17, 0.57]
IAXO = loadtxt("limit_data/IAXO.txt")
plt.plot(IAXO[:,0],IAXO[:,1],'--',linewidth=3,color=IAXO_col)

ADMX_Projected = loadtxt("limit_data/ADMX_Projected.txt")
plt.plot(ADMX_Projected[:,0],ADMX_Projected[:,1],'--',linewidth=3,color=ADMX_col,zorder=3)



# DM lines
DM_col = [0.25, 0.4, 0.96]
Axion_CDM = loadtxt("limit_data/Axion_CDM.txt")
ALP_CDM = loadtxt("limit_data/ALP_CDM.txt")
plt.plot(Axion_CDM[:,0],Axion_CDM[:,1],'-',linewidth=3,color=DM_col,zorder=3)
plt.plot(ALP_CDM[0:2,0],ALP_CDM[0:2,1],'-',linewidth=3,color=DM_col,zorder=3)
plt.plot(ALP_CDM[1:3,0],ALP_CDM[1:3,1],'--',linewidth=3,color=DM_col,zorder=3)


# Labels
plt.legend(loc='upper right',fontsize=15,frameon=False)
plt.xlabel(r"$m_a$ [eV]",fontsize=25)
plt.ylabel(r"$|g_{a\gamma}|$ [GeV$^{-1}$]",fontsize=25)
plt.text(5e-10,2e-7,r'{\bf ALPS-I}',fontsize=20,color='w')
plt.text(5e-10,2e-9,r'{\bf CAST+Sumico}',fontsize=20,color='w')
plt.text(1e1,1.5e-9,r'{\bf Solar} $\nu$',fontsize=20,color='w')
plt.text(200.5744,4.2720e-11,r'{\bf Ionisation}',fontsize=18,color='w',rotation=-90)
plt.text(70,3.2720e-11,r'{\bf fraction}',fontsize=18,color='w',rotation=-90)
plt.text(1e4,2e-14,r'{\bf EBL}',fontsize=25,color='w',rotation=-60)
plt.text(5e3,1e-16,r'{\bf X-rays}',fontsize=20,color='w',rotation=-60)
plt.text(3e-1,0.13e-11,r'{\bf Telescopes}',fontsize=12,color=Telescopes_col)
plt.text(0.6e0,1.3e-10,r'{\bf Globular clusters}',fontsize=18,color='w')
plt.text(2e-8,1.55e-11,r'{\bf HESS}',fontsize=13,color=HESS_col)
plt.text(4.8e-10,1.2e-11,r'{\bf Fermi}',fontsize=13,color='w')
plt.text(2.3e-11,1.5e-11,r'{\bf SN}-$\gamma$',fontsize=13,color='w')
plt.text(1.5e-12,2.5e-11,r'{\bf Hydra}',fontsize=11,color='w')
plt.text(1.5e-12,1.5e-11,r'\quad {\bf A}',fontsize=11,color='w')
plt.text(0.000000000001101,   0.000000000003946,r'\quad {\bf M87}',fontsize=11,color='w')
plt.text(0.7e-5,3e-11,r'{\bf RBF+UF}',fontsize=13,color='w',rotation=-90)
plt.text(4.1e-5,2.5e-12,r'{\bf HAYSTAC}',fontsize=14,color=HAYSTAC_col,rotation=-90)
plt.text(1.12e-6,1e-14,r'{\bf ADMX}',fontsize=15,color=ADMX_col,rotation=90)
plt.text(5e-3,2e-13,r'{\bf Axion}',fontsize=15,color=DM_col,rotation=-90)
plt.text(1.8e-3,2.5e-13,r'\quad {\bf CDM}',fontsize=15,color=DM_col,rotation=-90)
plt.text(1e-11,1.3e-13,r'{\bf ALP CDM}',fontsize=16,color=DM_col,rotation=32)
plt.text(0.3e-3,6e-12,r'{\bf IAXO}',fontsize=18,color=IAXO_col)
plt.text(3.9211e+03,3.3306e-11,r'{\bf Early Universe}',fontsize=20,rotation=-62,color='w')



# Limits
ax.set_xlim(left=m[0], right=m[-1])
ax.set_ylim(bottom=g[0], top=g[-1])

# Style
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
ax.tick_params(which='major',direction='in',width=2,length=10)
ax.tick_params(which='minor',direction='in',width=1,length=7)
plt.xscale('log')
plt.yscale('log')
plt.show()
fig.savefig('AxionLimits.pdf')
```
