# Axion-neutron coupling limits vs axion mass

Axion mass-PQ scale relation:
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

Axion-neutron coupling
\begin{equation}
g_{an} \equiv \frac{C_{an} m_{n}}{f_{a}} = 1.644\times 10^{-7} C_{an}\frac{m_a}{\mathrm{eV}}
\end{equation}

Model dependent constant:
\begin{equation}
C_{an}  = 
\begin{cases}
  -0.02(4) & {\rm KSVZ} \\
  [-0.16,0.26] & {\rm DFSZ\,I} \\
  [-0.16,0.26] & {\rm DFSZ\,II}
\end{cases}
\end{equation}
Like the axion-electron coupling the range of values for DFSZ I and II come from the perturbativity of the Yukawa couplings with sets the range $0.28<v_{u} / v_{d}<140$ for the Higgs vevs.

```python
%matplotlib inline
from PlotFuncs import BlackHoleSpins, AxionNeutron, MySaveFig

fig,ax = AxionNeutron.FigSetup(Shape='Rectangular',mathpazo=True)
AxionNeutron.QCDAxion(ax)
AxionNeutron.Haloscopes(ax,projection=False)
AxionNeutron.StellarBounds(ax)
AxionNeutron.LabExperiments(ax,projection=False)
BlackHoleSpins(ax,[3e-18,5e-17],alpha=0.1,PlotLine=False,rotation=0,text_col='dimgray')


from numpy import *


m_vals = array([1e-10,1e-2])
g_QCD_upper = 1.644e-7*0.26*m_vals
for i in logspace(1,8,40):
    ax.fill_between(m_vals,g_QCD_upper/i,y2=g_QCD_upper,color='orange',\
                    alpha=0.01,zorder=-100,lw=3)

MySaveFig(fig,'AxionNeutron')
```
