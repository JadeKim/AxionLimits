# Axion-photon coupling limits vs axion mass

Axion mass-PQ scale relation:
\begin{equation}
m_{a}=5.70(7) \mu \mathrm{eV}\left(\frac{10^{12} \,\mathrm{GeV}}{f_{a}}\right)
\end{equation}

Axion-photon coupling
\begin{equation}
g_{a \gamma} \equiv \frac{\alpha}{2 \pi} \frac{C_{a\gamma}}{f_{a}}=2.0 \times 10^{-16} C_{a \gamma} \frac{m_{a}}{\mu \mathrm{eV}} \mathrm{GeV}^{-1}
\end{equation}

Model dependent constant:
\begin{equation}
C_{a\gamma} \equiv \frac{E}{N} - 1.92(4) = 
\begin{cases}
  0.75, & {\rm KSVZ} \\
  -1.25, & {\rm DFSZ\,I} \\
  -1.92, & {\rm DFSZ\,II}
\end{cases}
\end{equation}

```python
%matplotlib inline
from numpy import *
import matplotlib.pyplot as plt
from PlotFuncs import AxionPhoton, MySaveFig

fig,ax = AxionPhoton.FigSetup(Shape='Rectangular',ylab='$|g_{a\gamma}|$ [GeV$^{-1}$]',mathpazo=True)
AxionPhoton.QCDAxion(ax)
AxionPhoton.Cosmology(ax)
AxionPhoton.StellarBounds(ax)
AxionPhoton.Haloscopes(ax,projection=False)
AxionPhoton.Helioscopes(ax,projection=False)
AxionPhoton.LSW(ax,projection=False)
AxionPhoton.AstroBounds(ax,projection=False)
AxionPhoton.NeutronStars(ax)


MySaveFig(fig,'AxionPhoton')
```
