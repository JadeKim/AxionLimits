```python
# Import astropy.io.fits for reading fits file
from astropy.io import fits
import matplotlib.pyplot as plt
from numpy import *

file = fits.open('24022020B.fits')
dat = file[1].data

#Get ra and dec columns
ra = dat['ra']
dec = dat['dec']

# Get pmra and pmdec columns
pmra = dat['pmra']
pmdec = dat['pmdec']

#Get G and BP-RP columns
G = dat ['phot_g_mean_mag']
BPRP = dat ['bp_rp']

#Get PM requirements
pmra_err = dat ['pmra_error']
pmdec_err = dat ['pmdec_error']

#Get astrometric fits requirements
astrochi = dat ['astrometric_chi2_al']
astrogoodobs = dat ['astrometric_n_good_obs_al']
u = (astrochi/astrogoodobs)**(1/2)

#Excess Flux Factor
E = dat['phot_bp_rp_excess_factor']

mask1 = (pmra<(0.2 + 2*pmra_err))&(pmdec<(0.2 + 2*pmdec_err))
mask2 = (u < 1.2* maximum(1,exp(-0.2*(G-19.5))))
mask3 = (E > (1 + 0.015*((BPRP)**2)))&(E < 1.5*(1.3 + 0.06*((BPRP)**2)))
mask4 = (-0.4 < BPRP)&(BPRP < 0.7)&(16 < G)&(G < 20)
mask5 = (22.1<(G+2.5*(BPRP)))&((G+2.5*(BPRP))<25.9)&(14.586<(G-(1.071*(BPRP))))&((G-(1.071*(BPRP)))<17.886)
```

```python
#Ellipse Cut #Equation for an ellipse = (x-x0)/a +(y-y0)/b
#Matrix of rotation = (x;y)(cos , sin ; -sin, cos) #Final equation = (x-x0)/a * cos(theta) + (x-x0)/a *sin(theta), (y-y0)/b *-sin(theta)+(y-y0)/b *cos(theta)
#Andromeda
x0=10.6847083
y0=41.26875
# Triangulum 23.4621,30.65994167
#x0 = 23.4621
#y0 = 30.65994167
x=ra
y=dec
a = 20000000000
#1.8
b = 0.4
#0.4
theta = (pi/180)*43
#43

mask_e = ((((x-x0)*cos(theta) + (y-y0)*sin(theta)/a)**2 + (((x-x0)*(-sin(theta)) + (y-y0)*cos(theta))/b)**2 ) < 1)
maskA = (ra<11.0)&(ra>10.5)&(dec<y0)
maskB = (ra<10.5)&(ra>10.)&(dec<(40.8))
maskC = (ra<10.)
maskB = (ra<10.5)&(ra>10.)&(dec<(40.8))
maskB = (ra<10.5)&(ra>10.)&(dec<(40.8))
maskB = (ra<10.5)&(ra>10.)&(dec<(40.8))
maskB = (ra<10.5)&(ra>10.)&(dec<(40.8))
maskE = (ra<11.0)&(ra>10.5)&(dec>y0)


#plt.scatter(ra[mask_e],dec[mask_e],s=1,c='grey')
plt.scatter(ra[mask_e*mask1*mask2*mask3*mask4],dec[mask_e*mask1*mask2*mask3*mask4],s=1,c='blue')
plt.scatter(ra[mask_e*mask1*mask2*mask3*mask5],dec[mask_e*mask1*mask2*mask3*mask5],s=1,c='red')
plt.scatter(ra[mask_e*mask1*mask2*mask3*maskA],dec[mask_e*mask1*mask2*mask3*maskA],s=1,c='green')
plt.scatter(ra[mask_e*mask1*mask2*mask3*maskB],dec[mask_e*mask1*mask2*mask3*maskB],s=1,c='grey')
plt.scatter(ra[mask_e*mask1*mask2*mask3*maskC],dec[mask_e*mask1*mask2*mask3*maskC],s=1,c='violet')
#plt.scatter(ra[mask_e*mask1*mask2*mask3*maskD],dec[mask_e*mask1*mask2*mask3*maskD],s=1,c='grey')
plt.scatter(ra[mask_e*mask1*mask2*mask3*maskE],dec[mask_e*mask1*mask2*mask3*maskE],s=1,c='skyblue')

pmtot = sqrt(pmra**2 + pmdec**2)
mean_pmra = mean(pmra)
mean_pmdec = mean(pmdec)
#pmgrey_avg = mean(pmtot[mask_e*mask1*mask2*mask3*mask6])
pmraA_avg = mean(pmra[mask_e*mask1*mask2*mask3*maskA])
pmdecA_avg = mean(pmdec[mask_e*mask1*mask2*mask3*maskA])
xpos1 = 10.75
ypos1 = 40.85
dx1 = pmraA_avg*4 - mean_pmra*4
dy1 = pmdecA_avg*4 -mean_pmdec*4
plt.arrow(xpos1, ypos1, dx1, dy1, head_width=0.06,length_includes_head=True)
pmraB_avg = mean(pmra[mask_e*mask1*mask2*mask3*maskB])
pmdecB_avg = mean(pmdec[mask_e*mask1*mask2*mask3*maskB])
xpos2 = 10.25
ypos2 = 40.4
dx2=pmraB_avg*4 - mean_pmra*4
dy2 = pmdecB_avg*4 - mean_pmdec*4
plt.arrow(xpos2,ypos2,dx2,dy2,head_width=0.06,length_includes_head=True)
pmraC_avg = mean(pmra[mask_e*mask1*mask2*mask3*maskC])
pmdecC_avg = mean(pmdec[mask_e*mask1*mask2*mask3*maskC])
xpos3 = 9.75
ypos3 = 40.45
dx3 = pmraC_avg*4 - mean_pmra*4
dy3 = pmdecC_avg*4 -mean_pmdec*4
plt.arrow(xpos3, ypos3, dx3, dy3, head_width=0.06,length_includes_head=True)
pmraE_avg = mean(pmra[mask_e*mask1*mask2*mask3*maskE])
pmdecE_avg = mean(pmdec[mask_e*mask1*mask2*mask3*maskE])
xpos5 = 10.75
ypos5 = 41.7
dx5=pmraE_avg*4 - mean_pmra*4 
dy5 = pmdecE_avg*4 - mean_pmdec*4
plt.arrow(xpos5,ypos5,dx5,dy5,head_width=0.06,length_includes_head=True)


print([dx1,dy1],[dx2,dy2],[dx5,dy5])
```
